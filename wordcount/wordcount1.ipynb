```python
from collections import defaultdict
import re
```

```python
# 替换除了n't这类连字符外的所有非单词字符和数字字符
def replace(s):
    if s.group(1) == 'n\'t':
        return s.group(1)
    return ' '
```

```python
def cal(filename='wordcount.txt'):
    # 使用lambda来定义简单的函数
    dic = defaultdict(lambda: 0)
    with open(filename, 'r') as f:
        data = f.read()
    # 全部变为小写字母
    data = data.lower()
    # 替换除了n't这类连字符外的所有非单词字符和数字字符
    data = re.sub(r'(n[\']t)|([\W\d])', replace, data)
    datalist = re.split(r'[\s\n]+', data)
    for item in datalist:
        dic[item] += 1
    del dic['']
    return dic
```

```python
if __name__ == '__main__':
    dic = cal()
    for key, val in dic.items():
        print('%15s  ---->   %3s' % (key,val))
```

```python
# 增加排序函数
def top_counts(dic, n=10):
    value_key_pairs = [(count, tz) for tz, count in dic.items()]
    value_key_pairs.sort()
    return value_key_pairs[-n:]
top_counts(dic)
```
